```python
def clean_text(a_text_block):
    a_text_block = a_text_block.encode('ascii', 'ignore').decode()
    a_text_block = a_text_block.replace('and/or',"and")
    
    a_text_block = a_text_block.replace('/'," or ")
    a_text_block = a_text_block.replace('etc'," ")
    a_text_block = a_text_block.replace('i.e',"that is")
    a_text_block = a_text_block.replace('e.g',"for example")
    #a_text_block = a_text_block.replace('.',"")
    a_text_block = a_text_block.replace('\"'," ")
    return a_text_block
```

```python
def chunks_to_token(chunk_tree):
    tokens=[]
    chunk_seq=False
    ch_token=''
    for tup, type_ in chunk_tree:

        if type_ == 'Chunk'and chunk_seq==True and tup[1] !='JJ':
            ch_token=ch_token+' '+tup[0]
        elif type_ == 'Chunk':
            chunk_seq=True
            ch_token=tup[0]
        else:
            if len(ch_token)>1:
                tokens.append((ch_token,'NN'))
                ch_token=''
            token = tup
            tokens.append(token)
            chunk_seq = False
    return tokens

            
          
        
```

```python
def reconstructed_tokens(a_text_block):
    a_text_block = clean_text(a_text_block)
    tokenized = nltk.word_tokenize(a_text_block)
    #print(tokenized)
    tagged = nltk.pos_tag(tokenized)
    chunkGram = r"""Chunk: {<JJ>*(<NN>|<NNS>)+}"""
    chunkParser = nltk.RegexpParser(chunkGram)
    chunked = chunkParser.parse(tagged)
    for subtree in chunked.subtrees(filter=lambda t: t.label() != 'Chunk'):
        #print(subtree.pos())
        chunk = subtree.pos()
    return chunks_to_token(chunk)
    
```
