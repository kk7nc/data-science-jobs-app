```python
jobs_with_states =jobs_skills.groupby(['posted_date','state'])['description'].count().astype(int)
jobs_with_states.head()
```

```python
states_df = jobs_with_states.reset_index()
```

```python
states_df.rename(columns={'description':'count'}, inplace = True)
```

```python
states_df['posted_date'] = states_df['posted_date'].apply(lambda x:dt.strptime(x,'%Y-%m-%d'))
```

```python
states_weekly = states_df.groupby([pd.Grouper(key='posted_date', freq='1w'),'state'])['count'].sum().astype(int)
```

```python
states_week = states_weekly.reset_index()
```

```python
top_states = states_week[states_week['state'].isin(['VA','WA','TX','NY','IL'])]
```

```python
top_states[top_states['state']=='VA']
```

```python
import seaborn as sns
sns.relplot(data=top_states,x='posted_date',y='count', hue='state',kind='line')
```
